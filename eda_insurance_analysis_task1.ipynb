```python
import pandas as pd

df = pd.read_csv('MachineLearningRating_v3.txt', delimiter='|')
```

```python
df.shape
df.columns
df.head()
```

```python
df.info()           # Check column types and nulls
df.describe()       # Summary statistics for numeric columns
df.isnull().sum()   # Count missing values per column
```

```python
# Drop fully missing columns
df.drop(columns=['NumberOfVehiclesInFleet', 'CrossBorder'], inplace=True)

# Convert dates
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
df['VehicleIntroDate'] = pd.to_datetime(df['VehicleIntroDate'], errors='coerce')  # coerce errors to NaT


df['CapitalOutstanding'] = pd.to_numeric(df['CapitalOutstanding'], errors='coerce')

# Fill categorical missing values with 'Unknown'
cat_cols = ['Bank', 'AccountType', 'MaritalStatus', 'Gender', 'NewVehicle',
            'WrittenOff', 'Rebuilt', 'Converted']
df[cat_cols] = df[cat_cols].fillna('Unknown')

# Fill numeric vehicle info with median
vehicle_num_cols = ['mmcode', 'Cylinders', 'cubiccapacity', 'kilowatts', 'NumberOfDoors']
df[vehicle_num_cols] = df[vehicle_num_cols].fillna(df[vehicle_num_cols].median())

df.drop(columns=['CustomValueEstimate'], inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt # Ensure matplotlib is imported

# Summary stats for numeric columns
df.describe()

# Check distribution of premiums and claims
# Capture the axes object
ax1 = df['TotalPremium'].hist(bins=50, figsize=(10, 4))
ax1.set_title('Total Premium Distribution') # Set the title using the axes object
ax1.set_xlabel('Total Premium')
ax1.set_ylabel('Frequency')
plt.show()

# Capture the axes object
ax2 = df['TotalClaims'].hist(bins=50, figsize=(10, 4))
ax2.set_title('Total Claims Distribution') # Set the title using the axes object
ax2.set_xlabel('Total Claims')
ax2.set_ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt # Ensure matplotlib is imported
import seaborn as sns # Import the seaborn library

# Scatter plot: Premium vs. Claims
df.plot.scatter(x='TotalPremium', y='TotalClaims', alpha=0.3, figsize=(8, 5), title='Premium vs Claims')

# Correlation matrix
correlation = df[['TotalPremium', 'TotalClaims', 'SumInsured', 'CalculatedPremiumPerTerm']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
